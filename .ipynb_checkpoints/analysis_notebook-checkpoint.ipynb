```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('data/Suicides_in_India_2001-2012.csv')
df.head()
```

### Cleaning Data

```python
modified_df=df[~df["State"].str.contains("Total")]
```

```python
states=modified_df['State'].replace(to_replace="A & N Islands", value="Andaman and Nicobar Islands")

states=states.replace(to_replace="D & N Haveli", value="Dadra and Nagar Haveli")

states=states.str.replace(r"\(.*\)","")
```

```python
modified_df["State"]=states
```

```python
modified_df.drop_duplicates(keep='first',inplace=True)
```

```python
modified_df.columns[modified_df.isna().any()].tolist()
```

```python
modified_df.to_csv('modified_data.csv',index=False)
```

```python
modified_df.head()
```

### Which age groups are most effected?  

```python
age_df=modified_df[['Age_group','Gender','Total']]
```

```python
#removing 0-100+ data as it isn't needed for the analysis
age_df=age_df[~age_df['Age_group'].str.contains('0-100+')]
```

```python
plt.figure(figsize=(10,10))
age_grp=age_df.groupby(['Age_group','Gender'],as_index=False).sum().sort_values('Total',ascending=False)
sns.barplot(x='Age_group',y='Total',hue='Gender',data=age_grp,palette='ocean_r')
plt.xlabel("Age Group",fontsize=16)
plt.ylabel("Frequency",fontsize=16)
plt.title("Suicide Rates w.r.t Age Groups",fontsize=21)
plt.tight_layout()
```

<h3><p>
Insights:<br /></h3>Most suicide rates are seen between the age of 15-44<br />
Mens peaks at age 30-44 whereas Women peaks at age 15-29<br />
ages of 0-14 are the lowest as expected<br />
</p>

### How does education play a role in this situation?

```python
edu_filt=modified_df['Type_code']=='Education_Status'
```

```python
education_df=modified_df[edu_filt][['Type','Gender','Total']]
```

```python
plt.figure(figsize=(10,10))
edu_grp=education_df.groupby(['Type','Gender'],as_index=False).sum().sort_values('Total',ascending=False)
sns.barplot(x='Type',y='Total',hue='Gender',data=edu_grp,palette='ocean_r')
plt.xticks(rotation=45,ha='right')
plt.xlabel("Last Qualification",fontsize=16)
plt.ylabel("Frequency",fontsize=16)
plt.title("Suicide Rates w.r.t Education Status",fontsize=21)
plt.tight_layout()
```

<h3><p>Insights:<br /></h3>
There is a clear trend as a person gets more educated the lesser are the suicide rates<br />
It should be noted that education may not directly correlate with education as being educated one generally acquires<br />
a stable career and being able to get educated to higher degrees shows they were financially sound from the start.
</p>

### How does Marital Status Effects this situatiion?

```python
agefilt=modified_df['Type_code']=='Social_Status'
```

```python
social_df=modified_df[agefilt][['Type','Gender','Total']]
```

```python
plt.figure(figsize=(10,10))
social_grp=social_df.groupby(['Type','Gender'],as_index=False).sum().sort_values('Total',ascending=False)
sns.barplot(x='Type',y='Total',hue='Gender',data=social_grp,palette='ocean_r')
plt.xlabel("Marital Status",fontsize=16)
plt.ylabel("Frequency",fontsize=16)
plt.title("Suicide Rates w.r.t Marital Status",fontsize=21)
plt.tight_layout()
```

<h3>Insights: <br /></h3>

to stay together.
</p>

### What are the other causes which are observed?

```python
causefilt=modified_df['Type_code']=='Causes'
```

```python
cause_df=modified_df[causefilt][['Type','Gender','Total']]
```

```python
#### Cleaning Cause data
```

```python
temp=cause_df['Type'].replace(to_replace="Bankruptcy or Sudden change in Economic", value="Bankruptcy")
temp=temp.replace(to_replace="Other Prolonged Illness", value="Prolonged Illness")
temp=temp.str.replace(r"\(.*\)","")
temp=temp.replace(to_replace="Illness ", value="Aids/STD")
temp=temp.replace(to_replace="Physical Abuse ", value="Rape and Abuse")
temp=temp.replace(to_replace="Dowry Dispute", value="Dowry")
temp=temp.replace(to_replace="Fall in Social Reputation", value="deteriorating Social Reputation")
temp=temp.replace(to_replace="Not having Children(Barrenness/Impotency", value="Not having Children")
temp=temp.replace(to_replace="Not having Children (Barrenness/Impotency", value="Not having Children")
```

```python
cause_df['Type']=temp
```

```python
maleFilt=cause_df['Gender']=="Male"
femaleFilt=cause_df['Gender']=="Female"
causeMale=cause_df[maleFilt][['Type','Total']]
causeFemale=cause_df[femaleFilt][['Type','Total']]
```

```python
space_series1=causeMale['Type'].str.replace(' ','_')
space_series2=causeFemale['Type'].str.replace(' ','_')
```

```python
causeMale['Type']=space_series1
causeFemale['Type']=space_series2
```

```python
plt.figure(figsize=(8,8))
causeMale_grp=causeMale.groupby(['Type'],as_index=False).sum().sort_values('Total',ascending=False)
sns.barplot(x='Type',y='Total',data=causeMale_grp,palette='twilight')
plt.xticks(rotation=45,ha='right')
plt.xlabel("Causes",fontsize=16)
plt.ylabel("Frequency",fontsize=16)
plt.title("Suicide Rates for Men w.r.t other Causes",fontsize=21)
plt.tight_layout()
```

<h3>Insights:<br /></h3>
<p>
Top causes for men to commit suicide are family problems and problems which they aren't comfortable sharing <br />
Bankruptcy is also a big cause for men which isn't the case for women. The reason may be that men are generally the<br />
sole bread winners for their families atleast in rural and orthodox families which increases the stress factor.
</p>

```python
plt.figure(figsize=(8,8))
causeFemale_grp=causeFemale.groupby(['Type'],as_index=False).sum().sort_values('Total',ascending=False)
sns.barplot(x='Type',y='Total',data=causeFemale_grp,palette='twilight')
plt.xticks(rotation=45,ha='right')
plt.xlabel("Causes",fontsize=16)
plt.ylabel("Frequency",fontsize=16)
plt.title("Suicide Rates for Women w.r.t other Causes",fontsize=21)
plt.tight_layout()
```

<h3>Insights:<br /></h3>
<p>
Top causes for women to commit suicide are family problems and problems which they aren't comfortable sharing <br />
Dowry is also a big cause for women which isn't the case for men. The reason may be that in rural india malpractices<br />
like dowry still happed causing stress and abuse to the women.
</p>

### State Wise Analysis

```python
state_df=modified_df[['State','Year','Total']]
```

<p>Top 4 States</p>

```python
state_top=state_df[['State','Total']]
statetop_grp=state_top.groupby('State')
statetop_grp.sum().sort_values(by='Total',ascending=False).head()
```

```python
topstatefilt=state_df['State'].eq("Maharashtra") | state_df['State'].eq("Tamil Nadu") | state_df['State'].eq("Andhra Pradesh") | state_df['State'].eq("Karnataka")
stateTopFour=state_df[topstatefilt]
state_grp=stateTopFour.groupby(['State','Year'])
state_grp.sum().sort_values('Total',ascending=False)
stateGrouped=state_grp.sum()
```

```python
x=stateTopFour['Year'].unique()
y_andhra=stateGrouped['Total'].iloc[0:12]
y_karna=stateGrouped['Total'].iloc[12:24]
y_maha=stateGrouped['Total'].iloc[24:36]
y_tamil=stateGrouped['Total'].iloc[36:48]
```

```python
plt.figure(figsize=(10,10))
plt.plot(x,y_andhra,label="Andhra Pradesh")
plt.plot(x,y_karna,label="Karnataka")
plt.plot(x,y_maha,label="Maharastra")
plt.plot(x,y_tamil,label="Tamil Nadu")
plt.legend()
plt.grid()
plt.xlabel("Top 4 States with Highest ",fontsize=16)
plt.ylabel("Frequency",fontsize=16)
plt.title("Suicide Rates for Women w.r.t other Causes",fontsize=21)
plt.tight_layout()
```

<h3>Insights:</h3><br />
<p>
The timespan of the data is observed from 2001-2012 <br />
Though with govenment sanctions and modernization of society I may had expected for the suicide rates with time but <br />
on the contrary it is seen that the rates has increased. This can be due to less awareness and not normalizing <br />
mental health treatment.
</p>

# Conclusion:
<ul>
<li>Compiling all the observations it is unfortunate to observe that with time suicide rates are on the rise and globalization and modernization most likely increased stress in the general public. The reason behind this stress can be the increase of competition and the social pressure of not being good enough.</li>
<li>Pre-teen and teenagers have the lowest suicide rates, where girls have a general high spike in this age gap then usual with prominent reasons being of sexual abuse and academic stress.</li>
<li>Boys and men are seen having the most cases at the age between 14-44. Most of the major life events like marriage, having kids, graduation, building a career occur in this timespan and men feel the stress of society and family depending on him to keep the family going is a major stress factor. Not being satisfied with salary is seen thus observing causes like bankrupcy, poverty and other money related problems</li>
<li>Farmer suicide is also a big factor and the steps taken by the government is clearly isn't enough.</li>
<li> Unhappy Mariage has came up as a big factor for both gender to commit suicide. </li>
</ul>

# Actions:
<ul>
<li>Spearding awareness of normalizing divorce is a big step to prevent suicides for both genders. Unhappy marriages/family is the reason for more than 50% people to commit suicide. This will be a big step and deciding factor of suicide prevention.</li>
<li>Farmers suicide is a major topic which needs to be addressed. Money is a major problem which can be address by making the loan sanction and repayment process easy. It must also be arranged that the famers get a justified marketplace where they can sell their crops at a respectable price.</li>
<li>Men suicide rates can be drastically reduced by normalizing the treatment of mental health issues and not treating the topic as a taboo. Though difficult for a third-world country like India but having a unemployment benefits where the eligible unemployed people can get a minimum salary will relieve alot of stress from men and in my asumptions will be much effective then the ration program the govenment currently has, though this assumption can only be validated with more data.</li>
<li>Women suicide rates are also highly preventable if they get a secured upbringing from childhood. Providing education and empower them to make a great career and choosing their own life patners if they choose to marry will be a big factor to save their lives.</li>
</ul>
